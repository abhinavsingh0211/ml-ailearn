```python
df=pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x=df['cgpa'], y=df['placement_exam_marks'], hue=df['placed'])
```

```python
from sklearn.linear_model import Perceptron
P=Perceptron()
```

```python
X = df[['cgpa', 'placement_exam_marks']]
Y = df['placed']
```

```python
X = df[['cgpa', 'placement_exam_marks']]
Y = df['placed']
print("X and Y have been correctly redefined.")
```

```python
P.fit(X, Y)
print("Perceptron fitted successfully.")
```

```python
y_pred = P.predict(X)
print("Predictions made by the Perceptron:")
print(y_pred)
```

Now, let's visualize the decision boundary of the perceptron.

```python
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10, 6))
plot_decision_regions(X.values, Y.values, clf=P, legend=2)
plt.xlabel('CGPA')
plt.ylabel('Placement Exam Marks')
plt.title('Perceptron Decision Boundary')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y, y_pred)
print(f"Accuracy of the Perceptron: {accuracy:.2f}")
```

```python
print(f"Accuracy of the Perceptron: {accuracy * 100:.2f}%")
```
